```
path ='image_file/1388-3.0.png'

img_array = cv2.imread(path ,cv2.IMREAD_GRAYSCALE)
new_array = cv2.resize(img_array, (100,100))  # resize t

im = cv2.resize(img_array, (28, 28))
im = [im.flatten()]

im = np.array(im)
im = im/255.0
pred = cf.predict(im)
if pred=="1":
  pred="Meningioma"
if pred=="2":
  pred="Glioma"
if pred=="3":
  pred="Pituitary Tumor"
plt.title(pred)
plt.imshow(new_array)
```
